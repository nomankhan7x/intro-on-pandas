```python
# prompt: import pandas library, read this file/content/Salaries.csv, can you apply functions like head , tail , type and attributes like shape, columns and also check for null values dataset

import pandas as pd

# Assuming Salaries.csv is in the /content directory
df = pd.read_csv('/content/Salaries.csv')

# Display the first 5 rows
print(df.head())

# Display the last 5 rows
print(df.tail())

# Get the shape (rows, columns)
print(df.shape)

# Get the column names
print(df.columns)

# Get the data type of the DataFrame
print(type(df))

# Check for null values (True/False)
print(df.isnull())

# Set option to display all rows (optional)
pd.set_option("display.max_rows", None)

# Display the DataFrame (with all rows if option is set)
print(df)

# Display the number of null values per column
print(df.isnull().sum())
```
